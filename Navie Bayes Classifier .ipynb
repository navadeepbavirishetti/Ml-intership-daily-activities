## Navie Bayes Classifier 

- Naive Bayes is a supervised machine learing algorithm, which means it learns from labeled data.
- It is mainly used for classification problems, such as spam va not spam ,play va not play
- the algorithm works using bayes' theorem , which help in calculating probabilities.
- naive bayes calculates the probability for each class.
- the class with the highest probability is selected as the final prediction.
- It is called "naive" because it assumes that all features are independent of each other, even though this is often not true in real life

## bayes' theorem 

- bayes' theorem  is a fundamental concept thet forms the basis of the naive bayes algorithm. it calculates the posterior probability of an evrnt occuring based on prior knowledege and evidence

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(r"C:\Users\keerthi\Downloads\weather_Navie_bayes_dataset.csv")
df
```

```python
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
df 
```

```python
x = df.drop("PlayTennis",axis=1)
y = df["PlayTennis"]
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("Predicted values:", y_pred)
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model accuracy:", accuracy)
```

```python
new_weather=pd.DataFrame(
    [[2, 0, 0, 1, 1]],
    columns= ["Day", "Outlook", "Temperature", "Humidity", "Wind"]
)
probability=model.predict_proba(new_weather) [0]
probability
```

```python
labels = ["No", "Yes"]
plt.figure()
plt.bar(labels, probability)
plt.xlabel("Play")
plt.ylabel("Probability")
plt.title("Naive Bayes Probability Graph (Play Tennis)")
plt.show()
```
