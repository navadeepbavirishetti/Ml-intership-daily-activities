## what is a data pipeline ?

- a data pipeline is a step-by-step process that takes raw data and convert it into model-ready data

## basic flow

- Raw Data → Cleaning → processing → model

## A data piprline:

- organizes data properly 
- automates steps
- reduces human errors
- makes ML models reliable

### pipeline in sklearn :

- A piprline in aklearn is a sequence of data preprocessing steps and a ML models combined to together to automate training and prediction
