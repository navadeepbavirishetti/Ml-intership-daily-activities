```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
     "age":[18,22,25,30,35,40,45,50]
})
print(data)
```

## equal-width

```python
data["age_bin"] = pd.cut(data["age"],bins=3)
data
```

```python
plt.hist(data["age"], bins=4)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal-width binning using pd.cut()")
plt.show()
```

```python
plt.hist(data["age"], bins=3)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal-width binning using pd.cut()")
plt.show()
```
